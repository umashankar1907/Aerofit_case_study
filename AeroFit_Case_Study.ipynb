# Remove/clip the data between the 5 percentile and 95 percentiles.

```python
data_1 = data


# Clipping outliers between 5th and 95th percentiles
percentiles = [5, 95]
for col in ['Age', 'Usage', 'Fitness', 'Income', 'Miles']:
    lower_limit, upper_limit = np.percentile(data_1[col], percentiles)
    data_1[col] = np.clip(data_1[col], lower_limit, upper_limit)

fig, axis = plt.subplots(nrows=3, ncols=2, figsize=(12, 10))
fig.subplots_adjust(top=1.2)

sns.boxplot(data=data_1, x="Age", orient='h', ax=axis[0,0])
sns.boxplot(data=data_1, x="Education", orient='h', ax=axis[0,1])
sns.boxplot(data=data_1, x="Usage", orient='h', ax=axis[1,0])
sns.boxplot(data=data_1, x="Fitness", orient='h', ax=axis[1,1])
sns.boxplot(data=data_1, x="Income", orient='h', ax=axis[2,0])
sns.boxplot(data=data_1, x="Miles", orient='h', ax=axis[2,1])
plt.show()
```

# 4.Representing the Probability

Find the marginal probability (what percent of customers have purchased KP281, KP481, or KP781)

```python
cross_tab_gender = pd.crosstab(index=data['Product'], columns=data['Gender'], margins=True, margins_name='Total')
print('Cross tab \n', cross_tab_gender)
marginal_probability_gender_product = (cross_tab_gender / cross_tab_gender.loc['Total', 'Total']) * 100
print('Percent of marginal probability \n',marginal_probability_gender_product)
```

```python
cross_tab_age = pd.crosstab(index=data['Product'], columns=data['Age'],margins_name='Total',margins=True)
print(cross_tab_age)
marginal_probability_age_product = (cross_tab_age / cross_tab_age.loc['Total', 'Total']) * 100
print('Percent of marginal probability \n',marginal_probability_age_product)
```

```python
cross_tab_Marital_status = pd.crosstab(index=data['Product'], columns=data['MaritalStatus'],margins_name='Total',margins=True)
print('Cross tab \n', cross_tab_Marital_status)
marginal_probability_Marital_status_product = (cross_tab_Marital_status / cross_tab_Marital_status.loc['Total', 'Total']) * 100
print('Percent of marginal probability \n',marginal_probability_Marital_status_product)
```

```python
cross_tab_Education = pd.crosstab(index=data['Product'], columns=data['Education'],margins_name='Total',margins=True)
print('Cross tab \n', cross_tab_Education)
marginal_probability_Education_product = (cross_tab_Education / cross_tab_Education.loc['Total', 'Total']) * 100
print('Percent of marginal probability \n',marginal_probability_Education_product)
```

```python
cross_tab_Fitness = pd.crosstab(index=data['Product'], columns=data['Fitness'],margins_name='Total',margins=True)
print('Cross tab \n', cross_tab_Fitness)
marginal_probability_Fitness_product = (cross_tab_Fitness / cross_tab_Fitness.loc['Total', 'Total']) * 100
print('Percent of marginal probability \n',marginal_probability_Fitness_product)
```

Find the probability that the customer buys a product based on each column.

```python
marginal_prob_product = pd.crosstab(index=data['Product'], columns='count', normalize=True)

# Calculating conditional probabilities for each combination of column values
conditional_prob_age = pd.crosstab(index=data['Age'], columns=data['Product'], normalize='index')
conditional_prob_gender = pd.crosstab(index=data['Gender'], columns=data['Product'], normalize='index')
conditional_prob_education = pd.crosstab(index=data['Education'], columns=data['Product'], normalize='index')
conditional_prob_fitness = pd.crosstab(index=data['Fitness'], columns=data['Product'], normalize='index')

print("Conditional Probability of Buying a Product based on Age:")
print(conditional_prob_age)

print("\nConditional Probability of Buying a Product based on Gender:")
print(conditional_prob_gender)

print("\nConditional Probability of Buying a Product based on Education:")
print(conditional_prob_education)

print("\nConditional Probability of Buying a Product based on Fitness:")
print(conditional_prob_fitness)
```

Find the conditional probability that an event occurs given that another even has occurred. (Example: given that a customer is female, what is the probability
she’ll purchase a KP481)

```python
# Probability of buying KP481 given that customer is a Female
# Calculating P(B)
marginal_prob_female = cross_tab_gender['Female'].values[1]

# Calculating the joint probability of buying KP481 and being female
joint_prob_female_product = (conditional_prob_gender.loc['Female'][1] * marginal_prob_product.loc['KP481']).values[0]

# Calculating P(A|B), the conditional probability of buying KP481 given that the customer is female
conditional_prob_product_given_female = (joint_prob_female_product / marginal_prob_female)*100

# Print the result
print(f"Conditional Probability of Buying KP481 given Female: {conditional_prob_product_given_female}")
```

```python
# Probability of buying KP481 given that customer is a Male
# Calculating P(B)
marginal_prob_male = cross_tab_gender['Male'].values[1]

# Calculating the joint probability of buying KP481 and being male
joint_prob_male_product = (conditional_prob_gender.loc['Male'][1] * marginal_prob_product.loc['KP481']).values[0]

# Calculating P(A|B), the conditional probability of buying KP481 given that the customer is male
conditional_prob_product_given_male = (joint_prob_male_product / marginal_prob_male)*100

# Print the result
print(f"Conditional Probability of Buying KP481 given Male: {conditional_prob_product_given_male}")
```
